```python
import pandas as pd
df = pd.read_csv('crimesdata.csv')
pop = pd.read_excel('chicagopop.csv')
```

```python
# Create a pivot table with the 'Area' as the index, the 'Primary Type' as the columns, and the 'ID' as the values
table2 = df.pivot_table(index='Area', columns='Primary Type', values='ID', aggfunc='count')

# Fill any missing values with 0
table2.fillna(0, inplace=True)   #cluster ανά Region 

```

```python
from sklearn import preprocessing 
df_standardized = preprocessing.scale(table2)
df_standardized = pd.DataFrame(df_standardized)
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(6,4))
wcss = []
for i in range (1, 9):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(df_standardized)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,9),wcss, 'bx-')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()  #θα κάνω κατευθείαν clustering με αριθμό 4 γιατί βγαίνουν τα καλύτερα αποτελέσματα μετά από αξιολόγηση
```

```python
kmeans = KMeans(n_clusters = 4, init = 'k-means++',random_state = 42)
y_kmeans = kmeans.fit_predict(df_standardized)
#beginning of the cluster numbering with 1 instead of 0 
y_kmeans1 = y_kmeans + 1

#list called cluster 
cluster = list(y_kmeans1)
#Adding cluster to our data set 
table2['cluster'] = cluster

#Mean of clusters 1 to 6 
kmeans_mean_cluster = pd.DataFrame(round(table2.groupby('cluster').mean(),1))
kmeans_mean_cluster.plot(kind='bar',stacked = True)
plt.rcParams["figure.figsize"] = (2,1)
plt.title("Cluster analysis")
plt.xlabel("Clusters")
```

```python
location = dff1[['Latitude','Longitude']]
```

```python
X = location.sort_values('Latitude').iloc[2:]
X = X[~np.isnan(X)]
```

```python
def doKmeans(X, nclust):
    model = KMeans(nclust)
    model.fit(X)
    clust_labels = model.predict(X)
    cent = model.cluster_centers_
    return (clust_labels, cent)

clust_labels, cent = doKmeans(X, 5)
kmeans = pd.DataFrame(clust_labels)
X.insert((X.shape[1]),'kmeans',kmeans)
```

```python
#Plot the clusters obtained using k means#Plot the 
fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(111)
scatter = ax.scatter(X['Longitude'],X['Latitude'],
                     c=kmeans[0],s=50)
ax.set_title('K-Means Clustering')
ax.set_xlabel('Long')
ax.set_ylabel('Lat')
plt.colorbar(scatter)
```

```python
import pandas as pd
df = pd.read_csv('crimesIN.csv')
```

```python
df
```

```python
from datetime import datetime

# extract hour from Date column and store it in a new column called 'Hour'
df['Hour'] = df['Date'].apply(lambda x: datetime.strptime(x, '%m/%d/%Y %I:%M:%S %p').hour)
df
```

```python
# extract month from Date column and store it in a new column called 'Month'
df['Month'] = df['Date'].apply(lambda x: datetime.strptime(x, '%m/%d/%Y %I:%M:%S %p').month)
df
```

```python
import datetime

def get_month_name(date):
    # Convert the date string to a datetime object
    date_object = datetime.datetime.strptime(date, '%m/%d/%Y %I:%M:%S %p')
    # Return the full month name
    return date_object.strftime('%B')

# Create the new 'Month Name' column
df['Month Name'] = df['Date'].apply(get_month_name)
```

```python
df
```

```python
df.groupby('Hour').count().sort_values('Month')
```

```python
#export to csv
df.to_csv('crimes.csv', index=False)
```
